```python
from image_lib import ImgLib as im
import os
import glob
import matplotlib.pyplot as plt
```

```python
dir = os.getcwd()
path_good_images = glob.glob(os.path.join(dir, 'Good/*.jpg'))
path_bad_images = glob.glob(os.path.join(dir, 'Bad/*.jpg'))
```

```python
img_good = im.read_image(path_good_images[0])
im.show_image(img_good)
```

```python
img_bad = im.read_image(path_bad_images[0])
im.show_image(img_bad)
```

```python
im.image_dimensions(img_good)
im.image_dimensions(img_bad)
```

```python
img_good_resized = im.resize_image(img_good, (512, 512))
img_gray = im.change_color_space(img_good_resized, 'GRAY')
im.show_image(img_gray)
```

```python
for i in range(6):
    im.draw_histogram(im.read_image(path_good_images[i]))
    im.draw_histogram(im.read_image(path_bad_images[i]))
```

### As We can see that for good images , the histogram is pretty smooth and for the bad images, the histogram is very jittery

### Let us check how does the red channel looks for both the images#


```python
hist = im.calculate_histogram(img_good, 1)
hist_ = im.calculate_histogram(img_bad, 1)

plt.plot(hist, color='red')
plt.plot(hist_, color="green") 
plt.show()

```

#### green denotes the red channel histogram for bad image and the red colored one is for the good image

```python
img_good_gray = im.change_color_space(img_good, 'GRAY')
img_bad_gray = im.change_color_space(img_bad, 'GRAY')
im.show_image(img_good_gray)
im.show_image(img_bad_gray)
```

```python
for i in range(6):
    im.draw_histogram(im.change_color_space(im.read_image(path_good_images[i]), 'GRAY'))
    im.draw_histogram(im.change_color_space(im.read_image(path_bad_images[i]), 'GRAY'), color='green')
```

### Green is for the bad images. For grayscale images, it is a bit difficult to identify which image will represent which category. Either way, we cannot just rely on intensity values.  But we can start the training with RGB images

### As in this case, finding gradients will not be able to help differentiate the classes. So Histogram of Gradients(HOG) as a feature extractor is also out of picture.
